### Decision Tree Classifier

```python
import pandas as pd 
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
from sklearn.datasets import load_iris
```

```python
dataset=load_iris()
```

```python
print(dataset.DESCR)
```

```python
dataset
```

```python
df=sns.load_dataset('iris')
df
```

```python
### dependent and dependent feature 
x=df.iloc[:,:-1]
y=df.iloc[::,-1]
```

```python
## x is the independent feature
x
```

```python
###  y is the independent feature
y
```

### splitting data set

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
### train the model 
from sklearn.tree import DecisionTreeClassifier

model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
from sklearn import tree 

plt.figure(figsize=(15,10))
tree.plot_tree(model,filled=True)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report

accuracy_score=accuracy_score(y_pred,y_test)
print(accuracy_score)
Report=classification_report(y_pred,y_test)
print(Report)
```

### Post Pruning with max_depth2 (Cut the branches)

```python
post_pruning=DecisionTreeClassifier(max_depth=2)
post_pruning.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(post_pruning,filled=True)
```

```python
y_pred=post_pruning.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report

accuracy_score=accuracy_score(y_pred,y_test)
print(accuracy_score)
Report=classification_report(y_pred,y_test)
print(Report)
```

### Decison tree with pre pruing And Hyperpparameter training For Huge Data

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
parameter={
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']

}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model_prepruning=DecisionTreeClassifier()
clf=GridSearchCV(model_prepruning,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
clf.fit(x_train,y_train)
```

```python
## Pre pruning parameters means best parameters 
clf.best_params_

```

```python
clf.predict(x_test)
```
